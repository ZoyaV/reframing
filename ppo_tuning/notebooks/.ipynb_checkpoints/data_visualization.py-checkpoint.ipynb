```python
import json
import os
import cv2
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import glob

img_path = glob.glob("../dataset/imgs/*")[1]
# Ð·Ð°Ð³Ñ€ÑƒÐ·Ð¸Ð¼ json Ñ„Ð°Ð¹Ð»
with open('../dataset/result.json', 'r') as f:
    data = json.load(f)

# Ð¿Ð¾Ð»ÑƒÑ‡Ð¸Ð¼ Ð²ÑÐµ ÑƒÐ½Ð¸ÐºÐ°Ð»ÑŒÐ½Ñ‹Ðµ category_id Ð² Ð´Ð°Ñ‚Ð°ÑÐµÑ‚Ðµ
unique_categories = set([d['category_id'] for d in data])

# ÑÐ¾Ð·Ð´Ð°Ð´Ð¸Ð¼ Ð²Ð¸Ð´Ð¶ÐµÑ‚ Ð´Ð»Ñ Ð²Ñ‹Ð±Ð¾Ñ€Ð° ÐºÐ°Ñ‚ÐµÐ³Ð¾Ñ€Ð¸Ð¹
category_selector = widgets.SelectMultiple(
    options=unique_categories,
    value=list(unique_categories),
    description='Categories',
    disabled=False
)
display(category_selector)

def plot_bboxes(image_name):
    # Ð·Ð°Ð³Ñ€ÑƒÐ·Ð¸Ð¼ Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸Ðµ
   # img_path = os.path.join( image_name)
    img = cv2.imread(image_name)
   # print(img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Ð½Ð°Ð»Ð¾Ð¶Ð¸Ð¼ bbox-Ñ‹ Ð½Ð° Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸Ðµ
    for item in data:
        print(item)
        if item['img_name'] == image_name:
            print(item['category_id'])
            if item['category_id'] in category_selector.value:
                print("Aboba")
                x, y, w, h = item['bbox']
                img = cv2.rectangle(img, (int(x), int(y)), (int(x+w), int(y+h)), (0, 255, 0), 2)
                cv2.putText(img, str(item['category_id']), (int(x), int(y-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

    # Ð²Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð¸Ñ€ÑƒÐµÐ¼ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚
    plt.figure(figsize=(12, 8))
    plt.imshow(img)
    plt.show()

def on_value_change(change):
    global img_path
    clear_output()
    display(category_selector)
    plot_bboxes(img_path)


category_selector.observe(on_value_change, 'value')

plot_bboxes(img_path)
```
